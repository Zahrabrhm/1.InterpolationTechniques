**Nearest Neighborhood Interpolation:**

Nearest Neighborhood Interpolation is one of the simplest interpolation methods used in image processing. When resizing or transforming an image, this method assigns the value of the nearest pixel to the interpolated point. In other words, it uses the color or intensity value of the closest existing pixel in the original image to represent the value of the new pixel. While this method is computationally efficient, it tends to produce blocky or pixelated results, especially when upscaling an image.

```python
import cv2
import matplotlib.pyplot as plt
# Load the images
cameraman = cv2.imread('./Data/cameraman.jpg')
house = cv2.imread('./Data/house.jpg')
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(cameraman, cv2.COLOR_BGR2RGB))
plt.title('Cameraman Original')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(house, cv2.COLOR_BGR2RGB))
plt.title('House Original')

plt.show()
```

```python

# Define the scaling factor
scale = 0.5  # Adjust this value according to your needs

# Apply nearest-neighbor interpolation
cameraman_resized = cv2.resize(cameraman, None, fx=scale, fy=scale, interpolation=cv2.INTER_NEAREST)
house_resized = cv2.resize(house, None, fx=scale, fy=scale, interpolation=cv2.INTER_NEAREST)

# Display the resized images using Matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(cameraman_resized, cv2.COLOR_BGR2RGB))
plt.title('Cameraman Resized')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(house_resized, cv2.COLOR_BGR2RGB))
plt.title('House Resized')

plt.show()
```

**Bilinear Interpolation:**

Bilinear Interpolation is a more sophisticated method that considers the values of the four nearest pixels surrounding the interpolated point. It calculates a weighted average of these pixel values based on the distance of the point from each pixel. This method results in a smoother transition between pixels compared to Nearest Neighborhood Interpolation. Bilinear Interpolation is commonly used in image scaling and rotation processes, providing better visual quality than Nearest Neighborhood Interpolation, especially for moderate changes in image size.

```python
# Define the scaling factor
scale = 0.5 # Adjust this value according to your needs

# Apply bilinear interpolation
cameraman_resized = cv2.resize(cameraman, None, fx=scale, fy=scale, interpolation=cv2.INTER_LINEAR)
house_resized = cv2.resize(house, None, fx=scale, fy=scale, interpolation=cv2.INTER_LINEAR)

# Display the resized images using Matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(cameraman_resized, cv2.COLOR_BGR2RGB))
plt.title('Cameraman Resized')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(house_resized, cv2.COLOR_BGR2RGB))
plt.title('House Resized')

plt.show()
```

**Bicubic Interpolation:**

Bicubic Interpolation is an extension of Bilinear Interpolation that involves considering a larger neighborhood of 16 surrounding pixels. Instead of linear interpolation, it employs a cubic polynomial to calculate the weighted average of pixel values. Bicubic Interpolation offers even smoother and more accurate results, making it suitable for applications where high-quality image interpolation is essential, such as in medical imaging or graphic design. While computationally more intensive than Nearest Neighborhood and Bilinear Interpolation, the improved visual quality often justifies its use in certain scenarios.

```python
# Define the scaling factor
scale = 0.5 # Adjust this value according to your needs

# Apply bicubic interpolation
cameraman_resized = cv2.resize(cameraman, None, fx=scale, fy=scale, interpolation=cv2.INTER_CUBIC)
house_resized = cv2.resize(house, None, fx=scale, fy=scale, interpolation=cv2.INTER_CUBIC)

# Display the resized images using Matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(cameraman_resized, cv2.COLOR_BGR2RGB))
plt.title('Cameraman Resized')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(house_resized, cv2.COLOR_BGR2RGB))
plt.title('House Resized')

plt.show()
```
